```python
from datasets import load_dataset
import pandas as pd
import os

os.makedirs("../data/raw", exist_ok=True)

# CSV list
csv_files = [
    "POS_CASH_balance.csv",
    "application_tr.csv",
    "application_ts.csv",
    "bureau.csv",
    "bureau_balance.csv",
    "credit_card_balance.csv",
    "installments_payments.csv",
    "previous_application.csv",
    "sample_submission.csv",
    "HomeCredit_columns_description.csv"
]

# User and repo in HuggingFace
USER = "jamirc"
REPO = "home_credit_default_risk"

for csv_name in csv_files:
    print(f"Downloading {csv_name} from HuggingFace...")

    # ðŸ‘‰ ESTA ES LA URL CORRECTA
    url = f"https://huggingface.co/datasets/{USER}/{REPO}/resolve/main/{csv_name}"

    ds = load_dataset("csv", data_files=url)
    df = ds["train"].to_pandas()

    output_path = f"../data/raw/{csv_name}"
    df.to_csv(output_path, index=False)

    print(f"{csv_name} saved: {len(df):,} rows, {len(df.columns)} columns")
    print("=" * 40)
```
